```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaler.fit(com.drop("Unnamed: 0", axis=1))

# Transform the data using the fitted scaler
X_c = pd.DataFrame(scaler.transform(com.drop("Unnamed: 0", axis=1)), columns=com.drop("Unnamed: 0", axis=1).columns)
```

```python
from sklearn.model_selection import train_test_split as tts
X_train,X_val = tts(X_c, test_size = 0.1, random_state=42)
```

```python
import tensorflow as tf
num_features = 768
latent_dim = 64 

model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(num_features,)),
    tf.keras.layers.Dense(512, activation="relu"), 
    tf.keras.layers.Dense(256, activation="relu"), 
    tf.keras.layers.Dense(latent_dim, activation="relu"), 
])

encoder = tf.keras.Model(inputs=model.input, outputs=model.layers[2].output)

decoder = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(latent_dim,)),
    tf.keras.layers.Dense(256, activation="relu"),
    tf.keras.layers.Dense(512, activation="relu"),
    tf.keras.layers.Dense(num_features, activation="sigmoid"),])

autoencoder = tf.keras.Model(inputs=model.input, outputs=decoder(encoder(model.input)))

class PrintLossCallback(tf.keras.callbacks.Callback):
   def on_epoch_end(self, epoch, logs=None):
       print(f"Epoch {epoch+1}: Training Loss: {logs['loss']:.4f}, Validation Loss: {logs['val_loss']:.4f}")

# Compile and train the autoencoder model
autoencoder.compile(loss="mse", optimizer="adam")
history = autoencoder.fit(
   X_train, X_train, epochs=5, batch_size=32,
   validation_data=(X_val, X_val),
   callbacks=[PrintLossCallback()]
) 

```

```python
percentiles = fulldata["IC50 (nM)"].describe(percentiles=[0.25, 0.75])
q1 = percentiles.loc['25%'] 
q3 = percentiles.loc['75%']

IQR = q3 - q1
lt = q1 - 1.5*IQR
ut = q3 +1.5*IQR

to_remove = []
for i in range(fulldata.shape[0]):
    if fulldata.iloc[i]["IC50 (nM)"]<lt or fulldata.iloc[i]["IC50 (nM)"]>ut:
        to_remove.append(i)
len(to_remove)
```

```python
fulldata = fulldata.reset_index()
fulldata = fulldata.drop('index', axis=1)
fulldata
```

```python
fulldata = fulldata.drop("Unnamed: 0", axis=1)
```

```python
fulldata = fulldata.drop(to_remove)
fulldata
```

```python
fulldata = fulldata.dropna()
fulldata.shape
```

```python
fulldata
```

```python
fulldata = fulldata.astype('float32')
```

```python
import pandas as pd

rows_to_drop = []

for column in fulldata.columns:   
    if fulldata[column].dtype == 'float32':       
        too_large_indices = fulldata.index[fulldata[column] > 3.4028235e+38].tolist()
                
        rows_to_drop.extend(too_large_indices)
# Drop rows with indices gathered
fulldata = fulldata.drop(rows_to_drop)
# Print the modified DataFrame
fulldata
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(fulldata.drop('IC50 (nM)', axis =1), fulldata['IC50 (nM)'], test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, HuberRegressor, ElasticNet, LogisticRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, ExtraTreesRegressor, AdaBoostRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import BayesianRidge, SGDRegressor
from sklearn.svm import SVR
from sklearn.metrics import r2_score
models = [
    ("RandomForest", RandomForestRegressor())]

for name, model in models:    
    model.fit(x_train, y_train)
    
    predictions = model.predict(x_test)    
    r2 = r2_score(y_test, predictions)
    print(f"{name:<25}: R^2 score = {r2:.4f}")
```

```python
acc = []
num = []
i_val = []
j_val = []
for k in range(0, 100):
    labels = []
    for i,j in zip(y_test, predictions):
        if i<=j+(k/2) and i>=j-(k/2):
            labels.append(1)
            if k<10:
                i_val.append(i)
                j_val.append(j)
        else:
            labels.append(0)
        
    acc.append((labels.count(1)/len(labels))*100)
    num.append(labels.count(1))
```
